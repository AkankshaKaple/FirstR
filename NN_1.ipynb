```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, accuracy_score
class MultiLinear:
    
    def __init__(self):
         # Learning Rate
        self.l_rate = 0.01
         # Total iterations
        self.epochs = 100
    
    
    def gradient_descent(self, x_train_data, y_train_data):
        x_train_data = x_train_data.T # (42 x 3165)
        y_train_data = y_train_data.T # (1 x 3165)
        w = np.random.rand(x_train_data.shape[0], 1)  # (42 x 1)
        b = np.zeros((1, 1), dtype='float')  # (1 x 1)
#         print("w", w.shape)
        

        m = len(x_train_data)
       
        for i in range(self.epochs):
            
            ######## Forward Propagation ########
            # Hypothesis function :
            z = (np.dot(w.T, x_train_data)) + b  # (1 x 42) * (42 x 3165) + 1 = (1 x 3165)

            # Activation Function :
            a = np.exp(z)/(1+np.exp(z))  # (1 x 3165)

#             # Loss Funtion :
#             J = np.dot(y_train_data.T, np.log(a)) - np.dot((1 - y_train_data.T), np.log(1 - a)) 

            ######## Backward Propagation ########
            dz = (a - y_train_data)  # (1 x 3165)   
#             print("x_train_data : ",x_train_data.shape)
#             print("dz : ", dz.shape)
            dw = (np.dot(x_train_data, dz.T)) / m  # (3165 x 1) * (3165 x 42) *  = (1 x 42)
            db = (np.sum(dz)) / m  # (1 x 1)

            # Gradient Descent :
            w = w - np.dot(self.l_rate, dw)  # (42 x 1)
            b = b - np.dot(self.l_rate, db)  # (1 x 1)
        return w, b
    
    # Prediction :
    def prediction(self, w, b, x_test_data):
        x_test_data = x_test_data.T
#         print("w : ", w.shape)
#         print("b : ", b.shape)
#         print("x_test_data : ", x_test_data.shape)
        z = (np.dot(w.T, x_test_data)) + b  # (1 x 42) * (42 x 3165) + 1 = (1 x 3165)
        return 1 / 1 + np.exp(-z)
    
          
          
def main():   
    import numpy as np
    import pandas as pd
    import matplotlib as plt
    from pandas.api.types import is_numeric_dtype
    import seaborn as sns

    df_1 = pd.read_csv("bank.csv", delimiter=";")
    df = pd.get_dummies(df_1,drop_first=True)
    for name in df.columns:
            if 'y' == name :     
                pass
            else:
                temp = 0
                temp_arr = np.array([])
                df[name] = (df[name] - df[name].mean()) / np.nanstd(df[name])

    
    x_data_set = np.array(pd.DataFrame(df, columns = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous',
           'job_blue-collar', 'job_entrepreneur', 'job_housemaid',
           'job_management', 'job_retired', 'job_self-employed', 'job_services',
           'job_student', 'job_technician', 'job_unemployed', 'job_unknown',
           'marital_married', 'marital_single', 'education_secondary',
           'education_tertiary', 'education_unknown', 'default_yes', 'housing_yes',
           'loan_yes', 'contact_telephone', 'contact_unknown', 'month_aug',
           'month_dec', 'month_feb', 'month_jan', 'month_jul', 'month_jun',
           'month_mar', 'month_may', 'month_nov', 'month_oct', 'month_sep',
           'poutcome_other', 'poutcome_success', 'poutcome_unknown']))

    y_data_set = np.array(pd.DataFrame(df, columns = ['y_yes']))


    train_size = int(np.ceil((len(y_data_set) * 70) / 100))
    test_size = int(len(y_data_set) - train_size)

    x_train_data = np.array(x_data_set[:train_size])
    y_train_data = np.array(y_data_set[:train_size])
    x_test_data = np.array(x_data_set[train_size:])
    y_test_data = np.array(y_data_set[train_size:])

    size = [x_train_data.shape]
    size = size[0][1] + 1

    m_l = MultiLinear()
#     print("x_train : ", x_train_data.shape)
#     print("y_data_train : ", y_train_data.shape)
#     print("x_data_test : ", x_test_data.shape)
#     print("y_test_data : ", y_test_data.shape)
    
    w, b = m_l.gradient_descent(x_train_data, y_train_data)
#     print("w1 : ", w.shape)
#     print("b1 : ", b.shape)
    y_prediction_test = m_l.prediction(w, b, x_test_data) 
    print("y_prediction_test : ", y_prediction_test.shape)

    y_prediction_train= m_l.prediction(w, b, x_train_data)  
    
  
    train_acc = round(float(100 - np.mean(np.abs(y_prediction_train - y_train_data)) * 100))

    test_acc = round(float(100 - np.mean(np.abs(y_prediction_test - y_test_data)) * 100))
    
    print("train accuracy:", train_acc)
    print("test accuracy:", test_acc)

if __name__ == '__main__':
    main()
```
